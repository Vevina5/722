```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('loanITERATION4.csv', encoding = "ISO-8859-1")
```

```python
# visualizing loan amount
#plt.hist(df.loan_amnt)
df.hist('loan_amnt')
plt.xlabel('(a) Loan Amount',fontsize=12)
plt.ylabel('Probability',fontsize=12)
plt.title('Histogram of Loan Amount',size=15)
plt.show()
```

```python
df.hist('funded_amnt',color='yellow')
plt.xlabel('(b) Funded Amount',fontsize=12)
plt.ylabel('Probability',fontsize=12)
plt.title('Histogram of Funded Amount',size=15)
plt.show()
```

```python
df.hist('int_rate', color = 'red')
plt.xlabel('(a) Interest Rate',fontsize=12)
plt.ylabel('Probability',fontsize=12)
plt.title('Histogram of Interest Rate',size=15)
plt.show()
```

```python
df.hist('installment', color='orange')
plt.xlabel('(b) Installment Amount',fontsize=12)
plt.ylabel('Probability',fontsize=12)
plt.title('Histogram of Installment Amount',size=15)
plt.show()
```
